```python
import pandas as pd
```

```python
df = pd.read_csv("customer_shopping_behavior.csv")
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount_usd'})
```

```python
df.columns
```

```python
labels = ['Young-Adult','Adult','Middle-Aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].head(20)
```

```python
frequency_mapping = {
    'Fortnightly'    : 14,
    'Weekly'         : 7,
    'Quarterly'      : 30,
    'Bi-Weekly'      : 14,
    'Annually'       : 365,
    'Monthly'       : 30,
    'Every 3 Months' : 90

}
df['purchases_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchases_frequency_days']].head(30)
```
